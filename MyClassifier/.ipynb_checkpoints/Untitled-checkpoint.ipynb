```python
import os
import struct
import random
import numpy as np
import matplotlib.pyplot as plt

def load_mnist(path, kind='train'):
    """Load MNIST data from `path`"""
    labels_path = os.path.join(path,
                               '%s-labels.idx1-ubyte'
                                % kind)
    images_path = os.path.join(path,
                               '%s-images.idx3-ubyte'
                               % kind)

    with open(labels_path, 'rb') as lbpath:
        magic, n = struct.unpack('>II',
                                 lbpath.read(8))
        labels = np.fromfile(lbpath,
                             dtype=np.uint8)

    with open(images_path, 'rb') as imgpath:
        magic, num, rows, cols = struct.unpack(">IIII",
                                               imgpath.read(16))
        images = np.fromfile(imgpath,
                             dtype=np.uint8).reshape(len(labels), 784)

    return images, labels

#### Loading the data

X_train, y_train = load_mnist('./data', kind='train')
X_test, y_test = load_mnist('./data', kind='t10k')
```

```python
C=0.1
eta=0.001
batch_size=500
num_iter=1
epsilon=0.001
```

```python
X_num, X_fea = np.shape(X_train)
print(X_num, X_fea)
```

```python
class_num=len(np.unique(y_train))
class_num
```

```python
w=np.random.rand(class_num,X_fea)
aver_w=w
w
```

```python
b=np.random.rand(class_num,1)
aver_b=b
b
```

```python
def shuffling(X, y):
        temp=list(zip(X,y))
        random.shuffle(temp)
        X,y=zip(*temp)
        return X,y
    
s_data, s_labels = shuffling(X_train, y_train)
```

```python
print(s_data)
```

```python
print(s_labels)
```

```python
def encoding(y):
        encoded_y=-1*np.ones((np.shape(y)[0],class_num))
        #encoded_y[60000][class_num]
        for i in range(np.shape(y)[0]):
            encoded_y[i,y[i]] = 1
        return encoded_y
    
encoded_y=encoding(s_labels)
```

```python
np.shape(encoded_y)
```

```python
cnt=0
(int)batch_count=X_num/batch_size
```

```python
def batching(X, y, t):                         
    batch_X=X[t*batch_size:min(len(X),(t+1)*batch_size)]
    #batch_X[batch_size][28*28]
    batch_y=y[t*batch_size:min(len(X),(t+1)*batch_size)]
    #batch_y[batch_size][class_num]
    last_size=min(len(X), (t+1)*batch_size)-t*batch_size
    #last_size[size][28*28]
    return batch_X, batch_y, last_size
```

```python
def net_input(X, w):
        #X[batch_size][28*28] w[class_num][28*28]
        net=np.dot(X,np.transpose(w))
        #[batch_size][class_num]
        return net
```

```python
def hinge_loss(X, y, w, b):
        net_v=net_input(X,w)
        #net_v[batch_size][class_num]
        temp=np.array(net_v)+np.transpose(b)
        #temp[batch_size][class_num]
        loss=y*temp
        #loss[batch_size][class_num]
        return loss
```

```python
for t in range(int(batch_count)):
    ###sgd
    batch_X, batch_y, bs=batching(s_data, encoded_y, t)
    #batch_X[batch_size][784] batch_y[batch_size][class_num] last_size=?
              
    batch_X=np.reshape(batch_X,(bs,X_fea))
    batch_y=np.reshape(batch_y,(bs,class_num))
                
    delta_w=np.zeros((class_num,X_fea))
    #delta_w[class_num][28*28]
    delta_b=np.zeros((class_num,1))
    #delta_b[10][1]
        
    loss=hinge_loss(batch_X, batch_y, w, b)
    #loss[batch_size][class_num]
    loss=1-loss-epsilon
    
    loss[loss<=0]=0
    loss[loss>0]=1
                
    l_M_y=loss*batch_y
    #l_M_y  =loss[bs][class_num] batch_y[bs][class_num]
        
    temp_w=np.dot(np.transpose(l_M_y),batch_X)
    #temp[10][28*28]=batch_y[batch_size][28*28]
    delta_w=-(1/bs)*np.array(temp_w)+(1/C)*np.array(w)
    #delta_w[10][28*28]=c*temp[10][28*28]+c*w[10][28*28]
                
    temp_b=np.sum(np.transpose(l_M_y),axis=1)
    temp_b=np.reshape(temp_b,(class_num,1))
    #temp[class_num][1]
    delta_b=-(1/bs) * temp_b
    #delta_b[10][1]=c*[class_num][1]
    cnt+=1
```

```python
delta_w
```

```python
delta_b
```

```python
for Xi in range(num_iter):
    w=np.array(w)-(eta*delta_w)
    #w[class_num][28*28]
    b=np.subtract(b,(eta * delta_b))
    #b[class_num][1]
            
    temp2_w=(cnt/cnt+1)*aver_w + (1/cnt+1)*w#
    temp2_b=(cnt/cnt+1)*aver_b + (1/cnt+1)*b#
            
    aver_w=np.where(aver_w>temp2_w,aver_w,temp2_w)
    aver_b=np.where(aver_b>temp2_b,aver_b,temp2_b)
```
